
Supervised learning involves training a model on a labeled dataset, where the algorithm learns the relationship between input features and corresponding output labels. There are various algorithms for supervised learning, and the choice of algorithm depends on the nature of the data and the problem at hand. Here's a list of commonly used supervised learning algorithms:

Linear Regression:

Used for regression tasks.
Assumes a linear relationship between input features and the target variable.


Logistic Regression:

Used for binary classification problems.
Estimates the probability that an instance belongs to a particular class.



Decision Trees:

Can be used for both classification and regression.
Hierarchical tree structure where each node represents a decision based on a feature.


Random Forest:

Ensemble learning method that combines multiple decision trees.
Can be used for both classification and regression tasks.


Support Vector Machines (SVM):

Effective for both classification and regression tasks.
Finds a hyperplane that best separates classes in the feature space.
k-Nearest Neighbors (KNN):



Instance-based learning algorithm.
Classifies a data point based on the majority class of its k nearest neighbors.


Naive Bayes:

Probabilistic algorithm based on Bayes' theorem.
Commonly used for classification tasks, especially in text classification.


Neural Networks:

Deep learning models with multiple layers (deep neural networks).
Suitable for complex tasks and large datasets.

Gradient Boosting Machines:

Ensemble learning method that builds a series of weak learners sequentially.
Examples include AdaBoost and Gradient Boosting Trees.


Linear Discriminant Analysis (LDA):

Used for dimensionality reduction and classification.
Assumes that the data are normally distributed and the classes have the same covariance matrix.


Ridge Regression:

A regularized linear regression method that adds a penalty term for large coefficients.
Helps prevent overfitting.


Lasso Regression:

Similar to Ridge Regression but uses a different regularization term.
Can lead to sparse models by causing some coefficients to be exactly zero.

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the California housing dataset
california_housing = fetch_california_housing()
data, target = california_housing.data, california_housing.target

# Use only one feature for simplicity - let's use the average rooms per household
# Reshape the data to meet the requirements of scikit-learn
X = data[:, np.newaxis, 5]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.2, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Plot the actual prices vs. predicted prices
plt.scatter(X_test, y_test, color='blue', label='Actual Prices')
plt.plot(X_test, predictions, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Average Rooms per Household')
plt.ylabel('House Value (in $100,000s)')
plt.title('Linear Regression on California Housing Data')
plt.legend()
plt.show()
```

polynomial regression data sets ->Polynomial regression is a type of linear regression where the relationship between the independent variable (X) and the dependent variable (Y) is modeled as an nth degree polynomial. This allows for a more flexible curve fitting compared to simple linear regression.

Let's consider a real-world example to illustrate polynomial regression. Suppose you are analyzing the relationship between the years of experience (X) an employee has and their salary (Y). In a simple linear regression, you might assume a linear relationship, but in reality, the relationship could be more complex.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection  import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error

california_housing=fetch_california_housing()

x=california_housing.data[:,np.newaxis,7]
y=california_housing.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)


poly=PolynomialFeatures(degree=2,include_bias=False)
x_train_poly=poly.fit_transform(x_train)

LR=LinearRegression()
LR.fit(x_train_poly,y_train)

y_train_pred=LR.predict(x_train_poly)


mse_error=mean_squared_error(y_train,y_train_pred)

print("The Mse value is {}",mse_error)

```

```python
plt.scatter(x_train,y_train,label="Traning of the data ")

x_range=np.linspace(x.min(),x.max(),100).reshape(-1,1)
x_range_poly=poly.transform(x_range)
y_range_pred=LR.predict(x_range_poly)
plt.plot(x_range,y_range_pred,color='red',label="Linear Regression")
plt.xlabel("Median Price")
plt.ylabel("Median House value ")
plt.title("Polynomila regression on california data sets ")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error

california_housing=fetch_california_housing()
data=pd.DataFrame(data=np.c_[california_housing['data'],california_housing['target']],columns=california_housing['feature_names']+['target'])

features = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude']
x=data[features]
y=data['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

elastic_net=ElasticNet(alpha=0.5,l1_ratio=0.5,random_state=42)

elastic_net.fit(x_train_scaled,y_train)

y_pred=elastic_net.predict(x_test_scaled)
print("The regression values are as follows {}".format(mean_squared_error(y_test,y_pred)))

print("Elastic coefficients is {}".format(elastic_net.coef_))
```

Logistic regression

Logistic regression is a statistical method used for binary classification problems. Unlike linear regression, which predicts a continuous output, logistic regression predicts the probability that an instance belongs to a particular category. The output of logistic regression is transformed using the logistic function (also called the sigmoid function), which maps any real-valued number to the range [0, 1]. This output can then be interpreted as the probability of belonging to the positive class.

The logistic regression model is trained by optimizing the parameters to maximize the likelihood of the observed data. This is typically done using optimization algorithms like gradient descent.

Logistic regression is widely used in various fields, including medicine, finance, and machine learning, for tasks where the goal is to predict a binary outcome.






```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,classification_report

iris=load_iris()
x=iris.data[:,:2]
y=(iris.target==2).astype(int)
scaler=StandardScaler()
x=scaler.fit_transform(x)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LogisticRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

accuracy = accuracy_score(y_test,y_pred)
classification_report=classification_report(y_test,y_pred)

print(f"Accuracy of the model : {accuracy:.2f} ")
print("classification report of the model :")
print(classification_report)

print(x)
```

A Decision Tree is a popular machine learning algorithm used for both classification and regression tasks. It's a tree-like model where an internal node represents a feature or attribute, the branch represents a decision rule, and each leaf node represents the outcome. The topmost node in a decision tree is known as the root node.

Here are the key concepts associated with Decision Trees:

1. **Root Node:**
   - The topmost node in the tree. It represents the feature that best splits the data based on a certain criterion.

2. **Internal Nodes:**
   - Nodes in the tree that are not leaf nodes. Each internal node represents a decision based on the value of a particular feature.

3. **Leaves:**
   - Terminal nodes at the end of the branches. Each leaf node represents the predicted outcome or class.

4. **Splitting:**
   - The process of dividing a node into sub-nodes based on the value of a particular feature. The goal is to increase the purity or homogeneity of the nodes.

5. **Decision Rule:**
   - The criteria used for splitting nodes, such as Gini impurity for classification or mean squared error for regression.

6. **Pruning:**
   - The process of removing branches from the tree to reduce overfitting and improve generalization to unseen data.

7. **Entropy and Gini Impurity:**
   - Measures of impurity used to evaluate the homogeneity of a set of labels. Decision Trees aim to minimize impurity in each node.

8. **Information Gain:**
   - A measure used to decide which feature to split on. It represents the reduction in entropy or Gini impurity achieved by a particular split.

Decision Trees are easy to understand and interpret, making them useful for visualizing and explaining decision-making processes. They can handle both numerical and categorical data and are capable of capturing complex relationships in the data. However, they are prone to overfitting, especially when the tree is deep. Techniques like pruning and setting a maximum depth help mitigate this issue.

Popular variations of Decision Trees include Random Forests, Gradient Boosted Trees, and CART (Classification and Regression Trees). These variations extend and improve upon the basic Decision Tree algorithm for different use cases.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report,accuracy_score

iris = load_iris()

x=iris.data
y=iris.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=DecisionTreeClassifier()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)

accuracy=accuracy_score(y_pred,y_test)
classificationreport=classification_report(y_pred,y_test)

print("The accuracy score is ",accuracy)
print("The classification report value is ",classificationreport)


plt.figure(figsize=(12,8))
plot_tree(model,feature_names=iris.feature_names,class_names=iris.target_names,filled=True,rounded=True)
plt.show()
```
